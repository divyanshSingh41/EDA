```python
def plot_graph(col_name):
  sns.displot(kind='hist', data=df, x=col_name, kde=True)
  sns.catplot(kind='box', data=df, x=col_name)
```

```python
cols = df.select_dtypes(include=['float64', 'int64']).iloc[:,[3, 4, 6, 9, 13, 14, 16]].columns
```

```python
for col in cols:
  plot_graph(col)
```
